```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16.0, 10.0)
```

```python
# Create plots with pre-defined labels.
fig, ax = plt.subplots()

g3 = ax.plot([0,777.77],[700,0], 'ro-', label="g3")
g2 = ax.plot([0,2250],[300,0], 'bo-', label="g2")
g1 = ax.plot([0,8000],[266.6,0], 'co-', label="g1")

legend = ax.legend(loc='top', shadow=True, fontsize='x-large')
# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('#00FFCC')
ax.legend(bbox_to_anchor=(1.1, 1.05))

plt.xlabel('lambda 1')
plt.ylabel('lambda 2')

plt.show()
```
